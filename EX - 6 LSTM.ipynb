```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.utils import to_categorical

# 1. Prepare text data
data = "hello world. this is a simple text generation example."
chars = sorted(list(set(data)))
char_to_idx = {c: i for i, c in enumerate(chars)}
idx_to_char = {i: c for i, c in enumerate(chars)}
vocab_size = len(chars)

# 2. Create sequences
X, y = [], []
for i in range(len(data) - 10):
    input_seq = data[i:i+10]
    target_char = data[i+10]
    X.append([char_to_idx[char] for char in input_seq])
    y.append(char_to_idx[target_char])

X = np.array(X)
y = np.array(y)

# 3. One-hot encode
X = to_categorical(X, num_classes=vocab_size)
y = to_categorical(y, num_classes=vocab_size)

# 4. Build and train the model
model = Sequential([
    LSTM(128, input_shape=(10, vocab_size)),
    Dense(vocab_size, activation='softmax')
])
model.compile(loss='categorical_crossentropy', optimizer='adam')
model.fit(X, y, epochs=100, verbose=1)

# 5. Generate text
def generate_text(seed_text):
    data = seed_text
    for _ in range(100):
        encoded = [char_to_idx[char] for char in data[-10:]]
        one_hot = to_categorical(encoded, num_classes=vocab_size).reshape(1, 10, vocab_size)
        next_index = np.argmax(model.predict(one_hot, verbose=0))
        data += idx_to_char[next_index]
    return data

# Example
seed_text = "hello worl"
print("Generated Text:\n", generate_text(seed_text))
```
